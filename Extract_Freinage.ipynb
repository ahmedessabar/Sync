# Automatisation Export Xsens (.mtb -> .txt)

**Version Corrig√©e (Copier-Coller)**
Ce script utilise le presse-papier pour coller le chemin du fichier, ce qui √©vite les erreurs de frappe.

**PR√â-REQUIS** :
1.  **MT Manager** doit √™tre **OUVERT** et visible.
2.  Ne **pas toucher** au PC pendant l'ex√©cution.

```python
# 1. INSTALLATION DES DEPENDANCES
%pip install pywinauto keyboard pyperclip
```

```python
# 2. SCRIPT D'EXPORTATION ROBUSTE
import time
from pathlib import Path
from pywinauto import Application, Desktop
from pywinauto.keyboard import send_keys
import pyperclip  # Pour g√©rer le copier-coller
import sys
import os

# --- CONFIGURATION ---
MBT_FOLDER = Path(r"c:\Users\es-sabar\Documents\PreTest\Moto_Freinage_mouille\Xsens")
EXPORT_TIMEOUT = 10 

# --- CONNEXION MT MANAGER ---
print("‚ñ∂ Connexion √† Xsens MT Manager (d√©j√† ouvert)...")

try:
    app = Application(backend="uia").connect(title_re=".*MT Manager.*", timeout=5)
    main_window = app.top_window()
    main_window.set_focus()
    print(f"   [OK] Connect√© √† : {main_window.window_text()}")
except Exception as e:
    print(f"‚ùå Erreur Connexion: {e}")
    print("   Fen√™tres visibles :")
    try:
        windows = Desktop(backend="uia").windows()
        for w in windows:
            t = w.window_text()
            if t: print(f"   - {t}")
    except: pass
    raise e

time.sleep(1)

# --- BOUCLE EXPORT ---
files = list(MBT_FOLDER.glob("*.mtb"))
print(f"Fichiers trouv√©s : {len(files)}")

for mbt_file in files:
    print(f"\nüìÇ Traitement : {mbt_file.name}")

    # 1. Ouvrir le dialogue (Ctrl+O)
    send_keys("^o")
    time.sleep(1)
    
    # 2. Copier le chemin complet dans le presse-papier
    full_path = str(mbt_file)
    pyperclip.copy(full_path)
    
    # 3. Coller (Ctrl+V) et Entr√©e
    send_keys("^v")
    time.sleep(0.5)
    send_keys("{ENTER}")
    
    time.sleep(5) # Attente chargement fichier

    # 4. Lancer Export (Ctrl+E)
    print("üì§ Lancement export...")
    send_keys("^e")
    time.sleep(2)
    
    # 5. Valider Export (Entr√©e)
    send_keys("{ENTER}")

    print("‚è≥ Attente fin export...")
    time.sleep(EXPORT_TIMEOUT)
    print(f"‚úÖ Export termin√© : {mbt_file.name}")

    # 6. Fermer le fichier (Ctrl+W)
    send_keys("^w")
    time.sleep(2)

print("\nüéâ Tous les fichiers ont √©t√© trait√©s.")
```
